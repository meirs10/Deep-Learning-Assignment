### 1. Sigmoid Function Gradient

#### 1.
$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$
$$
\sigma(x) = (1 + e^{-x})^{-1}
$$
$$
\frac{d\sigma}{dx} = -1 \cdot (1 + e^{-x})^{-2} \cdot \frac{d}{dx}(1 + e^{-x})
$$
$$
\frac{d\sigma}{dx} = -1 \cdot (1 + e^{-x})^{-2} \cdot (-e^{-x})
$$
$$
\frac{d\sigma}{dx} = \frac{e^{-x}}{(1 + e^{-x})^2}
$$

#### 2.
$$
\frac{d\sigma}{dx} = \left( \frac{1}{1 + e^{-x}} \right) \cdot \left( \frac{e^{-x}}{1 + e^{-x}} \right)
$$
$$
\frac{1}{1 + e^{-x}} = \sigma(x)
$$
$$
\frac{e^{-x}}{1 + e^{-x}} = \frac{(1 + e^{-x}) - 1}{1 + e^{-x}} = \frac{1 + e^{-x}}{1 + e^{-x}} - \frac{1}{1 + e^{-x}} = 1 - \sigma(x)
$$
$$
\frac{d\sigma}{dx} = \sigma(x)(1 - \sigma(x))
$$

### 2. Tanh Function Gradient

#### 1.
$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \quad \left[ \text{using } \frac{f'g - fg'}{g^2} \right]
$$
$$
f = e^x - e^{-x} \implies f' = e^x + e^{-x}
$$
$$
g = e^x + e^{-x} \implies g' = e^x - e^{-x}
$$
$$
\frac{d\tanh}{dx} = \frac{(e^x + e^{-x})(e^x + e^{-x}) - (e^x - e^{-x})(e^x - e^{-x})}{(e^x + e^{-x})^2}
$$
$$
\frac{d\tanh}{dx} = \frac{(e^x + e^{-x})^2 - (e^x - e^{-x})^2}{(e^x + e^{-x})^2}
$$

#### 2.
$$
\frac{d\tanh}{dx} = \frac{(e^x + e^{-x})^2}{(e^x + e^{-x})^2} - \frac{(e^x - e^{-x})^2}{(e^x + e^{-x})^2}
$$
$$
\frac{d\tanh}{dx} = 1 - \left( \frac{e^x - e^{-x}}{e^x + e^{-x}} \right)^2
$$
$$
\frac{d\tanh}{dx} = 1 - \tanh(x)^2
$$
